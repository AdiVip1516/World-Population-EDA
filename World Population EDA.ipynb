# World Population : Exploratory Data Analysis

In demographics, the term **World Population** is often used to refer to the total number of humans currently living, and was estimated to have exceeded 7.9 billion as of November 2021. It took over two million years of human prehistory and history for the human population to reach one billion and only 207 years more to grow to 7 billion. The global population is still increasing, but there is significant uncertainty about its long-term trajectory due to changing rates of fertility and mortality.

The total number of births globally is currently (2015–20) 140 million/year, is projected to peak during the period 2040-45 at 141 million/year and thereafter decline slowly to 126 million/year by 2100. The total number of deaths is currently 57 million/year and is projected to grow steadily to 121 million/year by 2100. The median age of human beings as of 2020 is 31 years.

Asia is the most populous continent, with its 4.64 billion inhabitants accounting for 60% of the world population. The world's two most populated countries, China and India, together constitute about 36% of the world's population. Africa is the second most populated continent, with around 1.34 billion people, or 17% of the world's population.

In this Notebook, EDA is done of **World Population** dataset recorded from 1970 to 2022, obtained from Kaggle.

## Importing the useful Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the World Population Dataset

```python
wp = pd.read_csv('world_population.csv')
wp.head(10)
```

## Sorting the dataset by Rank

```python
wp.set_index('Rank', inplace=True)
```

```python
wp.sort_index(inplace=True)
wp.head(10)
```

## Data Exploration

```python
wp.describe()
```

```python
wp.info()
```

## World Population Data of India

```python
wp_india = wp[wp['Country'] == 'India']
wp_india.head()
```

## Top 10 Most Populated Countries in 2022

```python
top_10_populated = wp.sort_values(['2022 Population','World Population Percentage'], ascending=False).head(10)
top_10_populated = top_10_populated.loc[:,['Country', '2022 Population', 'World Population Percentage']]
top_10_populated
```

### Plot

```python
top_10_populated.sort_values('World Population Percentage', ascending=False, inplace=True)
top_10_populated.plot(kind='bar',x="Country",y="2022 Population",figsize=(12,8),color='purple')
plt.title("Top 10 Most Populated Countries in 2022")
plt.ylabel("2022 Population (Billions)")
```

## World Population Percentage : Continent-Wise

```python
wp_continent_per = wp.loc[:, ['Continent', 'World Population Percentage']]
wp_continent_per = wp_continent_per.groupby('Continent').sum()
wp_continent_per
```

### Plot

```python
wp_continent_per.plot(kind='pie', y='World Population Percentage', figsize=(12,10), 
                      title='World Population Percentage : Continent-Wise')
```

## Top 10 Largest Countries by Area

```python
top_10_largest= wp.loc[:,['Country','Area (km²)']].sort_values('Area (km²)', ascending=False).head(10)
top_10_largest
```

### Plot

```python
top_10_largest.sort_values('Area (km²)', ascending=True, inplace=True)
top_10_largest.plot(kind='barh', x='Country', y='Area (km²)', color='indigo', figsize=(12,8))
plt.xlabel('Area (km²)')
plt.title('Top 10 Largest Countries by Area')
```

## Top 10 Smallest Countries by Area

```python
top_10_smallest= wp.loc[:, ['Country', 'Area (km²)']].sort_values('Area (km²)', ascending=True).head(10)
top_10_smallest
```

### Plot

```python
top_10_smallest.plot(kind='bar', x='Country', y='Area (km²)', color='orange', figsize=(12,8))
plt.xlabel('Country')
plt.ylabel('Area (km²)')
plt.title('Top 10 Smallest Countries by Area')
```

## Continent-Wise Area Distribution

```python
largest_continent_area= wp.loc[:,['Continent','Area (km²)']]
largest_continent_area = largest_continent_area.groupby('Continent').sum().reset_index()
largest_continent_area
```

### Plot

```python
largest_continent_area.plot(kind='bar', x='Continent', y='Area (km²)', color='seagreen', figsize=(12,8))
plt.xlabel('Continents')
plt.ylabel('Area (km²) per 10^7')
plt.title('Continent-Wise Area Distribution')
```

## Top 10 Highest Population Density by Country in 2022

```python
top_10_densed= wp.sort_values('Density (per km²)', ascending=False).head(10)
top_10_densed= top_10_densed.loc[:,['Country','Density (per km²)']]
top_10_densed
```

### Plot

```python
top_10_densed.sort_values('Density (per km²)', ascending=True, inplace=True)
top_10_densed.plot(kind='barh', x='Country', y='Density (per km²)', color='green', figsize=(12,8))
plt.xlabel('Density (km²)')
plt.title('Top 10 Highest Population Density by Country in 2022')
```

## Population Growth from 1970 to 2022 : Country-Wise

```python
wp['Population Growth'] = wp['2022 Population'] - wp['1970 Population']
wp['Population Growth (%)'] = (wp['Population Growth']/wp['1970 Population'])*100
wp_growth = wp.loc[:,['Country', '2022 Population', '1970 Population', 'Population Growth', 'Population Growth (%)']]
wp_growth
```

### Finding Country with the Maximum Population Growth

```python
print('Maximum Population Growth in Country : ', max(wp_growth['Population Growth (%)']))
wp_growth.sort_values(by=['Population Growth (%)'], ascending=False).head(1)
```

## Population Growth from 1970 to 2022 : Continent-Wise

```python
wp_growth_1 = wp.loc[:,['Continent', '2022 Population', '1970 Population', 'Population Growth']]
wp_growth_1 = wp_growth_1.groupby('Continent').sum().reset_index()
wp_growth_1['Population Growth (%)'] = (wp_growth_1['Population Growth']/wp_growth_1['1970 Population'])*100
wp_growth_1
```

### Finding Continent with the Maximum Population Growth

```python
print('Maximum Population Growth in Continent : ', max(wp_growth_1['Population Growth (%)']))
wp_growth_1.sort_values(by=['Population Growth (%)'], ascending=False).head(1)
```

## Plot showing World Population Difference between 1970 & 2022

```python
fig, ax = plt.subplots(figsize=(12,8))
plt.plot(wp_growth_1['Continent'], wp_growth_1['2022 Population'], label='2022', marker='o', color='red')
plt.plot(wp_growth_1['Continent'], wp_growth_1['1970 Population'], label='1970', marker='o', color='purple')
plt.grid(linewidth=0.4)
plt.title("World Population Difference between 1970 & 2022")
plt.xlabel('Continents')
plt.ylabel('Population (in Billions)')
plt.legend()
plt.show()
```

## Top 10 Countries with Highest Growth Rate

```python
top_10_growthrate = wp.loc[:, ['Country', 'Continent', '2022 Population', '1970 Population', 'Growth Rate']].sort_values(by='Growth Rate', ascending=False).head(10)
top_10_growthrate
```

### Plot

```python
fig, ax = plt.subplots(figsize=(12,8))
plt.plot(top_10_growthrate['Country'], top_10_growthrate['2022 Population'], label='2022', marker='o', color='green')
plt.plot(top_10_growthrate['Country'], top_10_growthrate['1970 Population'], label='1970', marker='o', color='magenta')
plt.xlabel('Country')
plt.ylabel('Growth Rate')
plt.grid(linewidth=0.4)
plt.title('Top 10 Countries with Highest Growth Rate')
plt.legend()
plt.show()
```

## Continent-Wise Growth Rate

```python
growth_rate_continent = wp.groupby('Continent')['Growth Rate'].sum().reset_index()
growth_rate_continent
```

### Plot

```python
growth_rate_continent.plot(kind='bar', x='Continent', y='Growth Rate', color='orange', figsize=(12,8))
plt.ylabel('Growth Rate')
plt.title('Growth Rate by each Continent')
plt.show()
```
